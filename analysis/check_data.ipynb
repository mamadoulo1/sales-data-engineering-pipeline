```python
# read csv files from data folder and print the first 5 rows of each file
import pandas as pd
import pathlib

path = pathlib.Path("C:/Users/mamad/Documents/Databricks/sales_data_engineering_pipeline/data/raw/customers.csv")
customers_df = pd.read_csv(path)
display(customers_df)
```

```python
# Drop duplicate rows from customers_df
customers_df = customers_df.drop_duplicates(subset=["customer_id"])
# Put signup_date column to datetime format
customers_df["signup_date"] = pd.to_datetime(customers_df["signup_date"], errors="coerce")
# FFill missing values in email to unknown@unknown.com
customers_df["email"] = customers_df["email"].fillna("unknown@unknown.com")
# Combine full name column from first_name and last_name columns and delete space with strip
customers_df["full_name"] = (customers_df["first_name"].str.strip() + " " + customers_df["last_name"].str.strip()).str.strip()
# Reste index of customers_df after dropping duplicates
customers_df = customers_df.reset_index(drop=True)
display(customers_df)
```

```python
# read orders.csv file and print the first 5 rows
path = pathlib.Path("C:/Users/mamad/Documents/Databricks/sales_data_engineering_pipeline/data/raw/orders.csv")
orders_df = pd.read_csv(path)
display(orders_df)
```

```python
# Find duplicate rows in orders_df using order_id column
duplicate_orders = orders_df[orders_df.duplicated(subset=["order_id"], keep=False)]
display(duplicate_orders)
```

```python
# Drop duplicate rows from orders_df
orders_df = orders_df.drop_duplicates(subset=["order_id"], keep="first")
display(orders_df)
```

```python
# Compter the number of unique order_id in orders_df
unique_order_count = orders_df["order_id"].nunique()
print(f"Number of unique order_id in orders_df: {unique_order_count}")
```

```python
# Find missing quantity values in orders_df
missing_quantity = orders_df[orders_df["quantity"].isnull()]
display(missing_quantity)
```

```python
# Fill missing quantity values with 1
orders_df["quantity"] = orders_df["quantity"].fillna(1)
# Change type of quantity column to int
orders_df["quantity"] = orders_df["quantity"].astype(int)
display(orders_df)
```

```python
# Parse order_date column to datetime format
orders_df["order_date"] = pd.to_datetime(orders_df["order_date"])
display(orders_df)
```

```python
invalid_dates = orders_df["order_date"].isna().sum()
print(f"Number of invalid dates in order_date column: {invalid_dates}")
```

```python
# Validate status column values
valid_statuses = ["pending", "shipped", "delivered", "cancelled"]
invalid_statuses = orders_df[~orders_df["status"].isin(valid_statuses)]
display(invalid_statuses)
```

```python
# Ensure that status column have only ["pending", "shipped", "delivered", "cancelled"
orders_df = orders_df[orders_df["status"].isin(valid_statuses)]
display(orders_df)
```

```python
# Fill missing discount values with 0.0
orders_df["discount_pct"] = orders_df["discount_pct"].fillna(0.0)
# Change type of discount column to float
orders_df["discount_pct"] = orders_df["discount_pct"].astype(float)
display(orders_df)
```

```python
# Reset index of orders_df after dropping duplicates and invalid statuses
orders_df = orders_df.reset_index(drop=True)
display(orders_df)
```

```python
# read products.csv file and print the first 5 rows
path = pathlib.Path("C:/Users/mamad/Documents/Databricks/sales_data_engineering_pipeline/data/raw/products.csv")
products_df = pd.read_csv(path)
display(products_df)
```

```python
# Drop duplicate rows from products_df
products_df = products_df.drop_duplicates(subset=["product_id"])
# Change unit_orice column to numeric type 
products_df["unit_price"] = pd.to_numeric(products_df["unit_price"], errors="coerce")
#drop rows with missing unit_price values
products_df = products_df.dropna(subset=["unit_price"])
# Reset index of products_df after dropping duplicates and missing values
products_df = products_df.reset_index(drop=True)
display(products_df)
```

```python
# Join orders, customers, products dataframes on customer_id and product_id columns
merged_df = orders_df.merge(customers_df, on="customer_id", how="left", validate="m:1").merge(products_df, on="product_id", how="left", validate="m:1")
display(merged_df)
```

```python
# Computing  financial metrics like gross_revenue, discount_amount, net_revenue, estimated_cost, gross_profit, order_year_month
merged_df["gross_revenue"] = merged_df["quantity"] * merged_df["unit_price"]
merged_df["discount_amount"] = merged_df["gross_revenue"] * merged_df["discount_pct"]
merged_df["net_revenue"] = merged_df["gross_revenue"] - merged_df["discount_amount"]
merged_df["estimated_cost"] = merged_df["gross_revenue"] * (1 - 0.3)
merged_df["gross_profit"] = merged_df["net_revenue"] - merged_df["estimated_cost"]
merged_df["order_year_month"] = merged_df["order_date"].dt.to_period("M").astype(str)
display(merged_df)
```

```python
# Build sales summary based on order_year_month and category columns with metrics like total_orders, total_quantity, gross_revenue, net_revenue, gross_profit, avg_discount_pct
sales_summary = merged_df.groupby(["order_year_month", "category"]).agg(
    total_orders=("order_id", "count"),
    total_quantity=("quantity", "sum"),
    gross_revenue=("gross_revenue", "sum"),
    net_revenue=("net_revenue", "sum"),
    gross_profit=("gross_profit", "sum"),
    avg_discount_pct=("discount_pct", "mean")
).reset_index()
display(sales_summary)
```

```python
# Crate an histogrm of profit_margin_pct column in sales_summary
```

```python
# Calculate profit margin percentage and add it as a new column to sales_summary a,d round it to 2 decimal places
sales_summary["profit_margin_pct"] = (sales_summary["gross_profit"] / sales_summary["gross_revenue"] * 100).round(2)
display(sales_summary)
```
